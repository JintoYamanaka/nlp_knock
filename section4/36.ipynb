```python
from mymodule import create_morphemes
```

```python
morphemes = create_morphemes()
morphemes[:5]
```

```python
words = []
for morpheme in morphemes:
    for line in morpheme:
        words.append(line["surface"])

words
```

```python
import collections

words_count = collections.Counter(words)
words_count
```

```python
sorted_words_count = words_count.most_common()
sorted_words_count
```
