```python
from mymodule import create_morphemes
from mymodule import create_words
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
morphemes = create_morphemes()
words = create_words(morphemes)
words[:10]
```

```python
'''
matplotlibで散布図
matplotlib.pyplot.scatter(x, y)
x, y：グラフに出力するデータ

他にも引数は多くある、随時チェック
'''

x = [i for i in range(1, len(words)+1)]
y = [word[1] for word in words]

plt.rcParams["font.size"] = 13
plt.scatter(x, y)
plt.title("39. Zipfの法則")
plt.xlabel("出現頻度（回数）")
plt.ylabel("出現頻度順位")
# 対数グラフに
plt.xscale('log')
plt.yscale('log')
plt.show
```
