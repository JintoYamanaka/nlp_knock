```python
from mymodule import create_morphemes
```

```python
morphemes = create_morphemes()
morphemes[:5]
```

```python
noun_phrase = []
for morpheme in morphemes:
    for i in range(1, len(morpheme)-2): # 最初と最後（句点の１つ前）の要素は調べる必要がないから1から始めて len(morpheme)-2で終わる
        if morpheme[i]["surface"] == "の" \
        and morpheme[i-1]["pos"] == "名詞" \
        and morpheme[i+1]["pos"] == "名詞":
            noun_phrase.append(morpheme[i-1]["surface"] + morpheme[i]["surface"] + morpheme[i+1]["surface"])

noun_phrase
```

```python
unique_noun_phrase = set(noun_phrase)
unique_noun_phrase
```
